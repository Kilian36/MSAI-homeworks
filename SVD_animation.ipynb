```python
import matplotlib.pyplot as plt
%matplotlib widget
import numpy as np
```

```python
import skimage.data as skd
img = skd.coins()
plt.imshow(img, 'gray')
plt.show()
img = img - np.mean(img)

U, s , VT = np.linalg.svd(img, full_matrices=False)

print('Shape of the SVD matrices non complete:\nU-> {}\nV-> {}'.format(U.shape, VT.shape))
```

```python
from matplotlib.animation import FuncAnimation
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(10, 6)
def animate(rg):
    Sum = 0
    for i in range(rg):
        Ui = U[:, i].reshape(len(U[0]), 1)
        VTi = VT[i, :].reshape(1, len(VT[0])) 
        Sum += Ui @ VTi * s[i]
    ax[0].clear()
    ax[0].imshow(Sum, cmap='gray')
    ax[0].set_title('{}-rank version of image'.format(i))

ax[1].imshow(img, cmap='gray')
anim = FuncAnimation(fig, animate, frames=range(1, 100), interval=100, repeat=False)
plt.show()
```
